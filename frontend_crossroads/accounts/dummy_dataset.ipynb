```python
# Create dummy data for account table
import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime

nr_of_customers = 1000 # the number of rows that are going to be generated

fake = Faker()

accounts = []

for customers_id in range(nr_of_customers):

    # Create transaction date 
    #d1 = datetime.strptime(f'1/1/2021', '%m/%d/%Y')
    #d2 = datetime.strptime(f'8/10/2021', '%m/%d/%Y')
    #transaction_date = fake.date_between(d1, d2)

    #create customer's name
    first_name = fake.first_name()
    last_name = fake.last_name()
    # Create account type
    account_type = random.choice(["B", "S"])

    # Create email 
    email = fake.ascii_email()

    #create account ID in 8-digit barcode
    account_ID = fake.ean(length=8)
    
    #create order id 
    order_ID = fake.ean(length=8)

    #create delivery_date
    delivery_date = datetime.strptime(f'1/1/2021', '%m/%d/%Y')

    #create address id
    address_id = fake.ean(length=8)

    #create address title
    address_title = fake.color_name()

    #create address street name
    addr_street_name = fake.street_address()
    # create address unit
    addr_unit_num = fake.building_number()
    #create address unit state
    addr_state = fake.state()

    #create address zipcode
    addr_zipcode = fake.postcode()

    #create review id
    review_id = fake.ean(length=8)

    #create item_id
    item_id = fake.ean(length=8)

    #create review score
    review_score = fake.random_digit()

    #create review title
    review_title = fake.word()

    #create review body
    review_body = fake.paragraph(nb_sentences=5)

    #create item title
    item_title = fake.sentence(nb_words=2)

    # create item price
    item_price = fake.pyint(max_value=500)
    #create item desc
    item_desc = fake.paragraph(nb_sentences=2)

    #create item_quantity
    item_quantity = fake.random_digit()
    
    #amount_spent = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=100)

    accounts.append([first_name, last_name, account_type, email, account_ID, order_ID, delivery_date,
                     address_id, address_title, addr_street_name, addr_state, addr_zipcode,
                     review_id,item_id,review_score,review_title,review_body,item_title,item_price,
                     item_desc, item_quantity])

accounts_df = pd.DataFrame(accounts, columns=['Account_fname', 'Account_lname','Account_type','Email', 'account_ID','order_ID','delivery_date',
                                              'address_ID','address_title','addr_street_name','addr_state','addr_zipcode',
                                              'review_id','item_id','review_score','review_title','review_body','item_title',
                                              'item_price','item_desc','item_quantity']) 
                
pd.pandas.set_option('display.max_columns', None)
print(accounts_df)
```
